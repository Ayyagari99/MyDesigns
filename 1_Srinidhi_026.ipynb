**QUESTION 4:** Write a python program to find the square root of a given number (n) within a given precision (p). Do not use any predefined library. (Take a guess, compute error at each successive step, until error is less than given precision) Example values of n are 10,15, 45, etc. Example values of p are 0.1, 0.05, 0.01, etc. [2 pts]

```
def squareRoot(n, p): 
  
    i = 0
    j = n
    root=1
  
    while (i <= j) : 
        mid = int((i + j) / 2) 
          
        if (mid * mid == n) : 
            root = mid 
            break
          
       #on right
        if (mid * mid < n) : 
            i = mid + 1
                    
        # on left 
        else : 
            j = mid - 1
          
    increment = 0.1
    for x in range(p):  
        while (root * root <= n): 
            root += increment 
          
        root = root - increment 
        increment = increment / 10
      
    return root 

print(squareRoot(90, 7)) 
```

And then for fractional part, upto p places.

**QUESTION 5:** Find the minimum value of a function y = (x+3)2. Write a python program to find the minimum value, use the numerical method described in this link. [2 pts]

```
x = 1 # The algorithm starts at x=1
alpha = 0.01 
precision = 0.000001 
s = 1 
max_i = 10000 # maximum iterations
i = 0 #counter
df = lambda x: 2*(x+3) #Gradient of our function 
```

At the beginning x=1, alpha= 0.01 (learning rate), precision= 0.000001.

When difference between two concsecutive x is less than precision/ max number of iterations done, algorithm stops. 

x*k+1*= x*k* - (learning rate)* (dy/dx)

```
import math

while s > precision and i < max_i:
    prev_x = x 
    x = x - alpha * df(prev_x) #Gradient descent
    s = abs(x - prev_x) 
    i = i+1 
    print("Iteration",i,"\nX value is",x) 
    
x= math.floor(x)
print("The local minimum occurs at", x)
print("The local minimum is ", (x+3)**2)
```
